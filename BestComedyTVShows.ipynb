```python
import requests
r = requests.get('https://editorial.rottentomatoes.com/guide/best-comedy-shows-of-all-time/')
print(r.text[0:1000])
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, 'html.parser')

results = soup.find_all('div', attrs = {'class': 'info synopsis'})
print(len(results))
print(results[-1])
```

```python
names = []
links = []
oResults = soup.find_all('div', attrs = {'class': 'article_movie_title'})
print(len(oResults))
# print(oResults[-1])
for oResult in oResults:
    nL = oResult.find('h2').find('a')
    name = oResult.find('h2').find('a').text
    link = oResult.find('h2').find('a')['href']
    names.append(name)
    links.append(link)
    # print(name)
    # print(link)
print("end")
print(names.pop(42))
print(links.pop(42))
```

```python
descs = []
for result in results:
    desc = result.find('span', class_= 'descriptor').next_sibling.strip()
    descs.append(desc)
```

```python
records = []
print(len(descs))
print(len(names))
print(len(links))
for i in range(49):
    records.append((names[i], descs[i], links[i]))
    print(records[i])
    print('\n')
```

```python
import pandas as pd
df = pd.DataFrame(records, columns=['names', 'descriptions', 'links'])
```
